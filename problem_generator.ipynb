# DSA Problem Generator Pipeline

This notebook connects to the backend API which integrates with the Google Gemini AI model to programmatically generate an unlimited stream of type-wise DSA problems.

```python
import requests
import json
import time

API_URL = "http://localhost:3000/problems/generate"
TOPICS = ["Arrays", "Dynamic Programming", "Graphs", "Trees", "Backtracking"]
DIFFICULTIES = ["Easy", "Medium", "Hard"]
```

```python
def generate_problems(topic, difficulty, count=3):
    problems = []
    for i in range(count):
        print(f"Generating {difficulty} {topic} problem ({i+1}/{count})...")
        try:
            response = requests.post(API_URL, json={"topic": topic, "difficulty": difficulty})
            response.raise_for_status()
            problem = response.json()
            problems.append(problem)
            print(f"Successfully generated: {problem.get('title')}")
        except Exception as e:
            print(f"Failed to generate: {e}")
        time.sleep(2) # Rate limiting
    return problems
```

```python
# Generate a batch of problems and save to a JSON seed file
all_generated_problems = []

for topic in TOPICS:
    all_generated_problems.extend(generate_problems(topic, "Medium", count=1))

with open('ai_problem_seed.json', 'w') as f:
    json.dump(all_generated_problems, f, indent=4)
    
print("Finished generating batch!")
```
